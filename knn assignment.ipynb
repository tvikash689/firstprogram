1. What is K-Nearest Neighbors (KNN) and how does it work.

ans.The K-Nearest Neighbors (KNN) algorithm is a simple machine learning method that predicts the class or value of a new data point based on the values of its nearest neighbors in the training data.

It works by finding the k closest data points in the training set to the new data point, and then using these neighbors to make a prediction.

2.What is the difference between KNN Classification and KNN Regression.

ans.Regression problems use a similar concept as classification problem, but in this case, the average the k nearest neighbors is taken to make a prediction about a classification. The main distinction here is that classification is used for discrete values, whereas regression is used with continuous ones.

3.What is the role of the distance metric in KNN.

ans.n K-Nearest Neighbors (KNN), the distance metric is crucial for determining the similarity or dissimilarity between data points. It defines how the algorithm measures the proximity between a query point and other data points in the training set, which is essential for identifying the K-nearest neighbors.

4. What is the Curse of Dimensionality in KNN.

ans.The Curse of Dimensionality in K-Nearest Neighbors (KNN) refers to the phenomenon where performance degrades as the number of features (dimensions) in the dataset increases.

5. 5 How can we choose the best value of K in KNN

ans.using hyperparameter tuning

6.What are KD Tree and Ball Tree in KNN.

ans.The tree is built by recursively dividing the data into two subsets, using a hyperplane perpendicular to one of the dimensions. This partitioning continues until each node represents a region containing a small number of data points.

7. When should you use KD Tree vs. Ball Tree.

ans.Slower than KD-Trees in low dimensions (d≤3) but a lot faster in high dimensions. Both are affected by the curse of dimensionality, but Ball-trees tend to still work if data exhibits local structure (e.g. lies on a low-dimensional manifold).

8. What are the disadvantages of KNN.

ans.KNN (k-Nearest Neighbors) has several disadvantages, including computational cost, memory usage, sensitivity to irrelevant features, and the curse of dimensionality

9.How does feature scaling affect KNN.

ans.Feature scaling significantly impacts KNN by ensuring all features contribute equally to distance calculations, which is crucial for accurate neighbor identification and classification. Without scaling, features with larger values might dominate the distance calculations, skewing the results and potentially leading to inaccurate classifications

10. What is PCA (Principal Component Analysis)

11. How does PCA work.

ans.PCA takes a dataset with multiple variables as input, and it produces a dataset into a lower subspace, that is, a reduced dataset with fewer variables. It is often used in exploratory data analysis for building predictive models, but it is also used in data preprocessing for dimensionality reduction

12. What is the geometric intuition behind PCA.

ans.PCA finds directions (principal components) in the data that capture the most variance, essentially transforming the data into a lower-dimensional space while preserving the most important information. Think of it as rotating the axes of your data to find the direction where the data is most spread out.


13. What is the difference between Feature Selection and Feature Extraction.

Feature selection and feature extraction are both used in machine learning to improve model performance, but they do so in different ways. Feature selection focuses on selecting a subset of the original features, while feature extraction creates new features from existing ones

14.What are Eigenvalues and Eigenvectors in PCA.

ans.In Principal Component Analysis (PCA), eigenvalues represent the amount of variance explained by each principal component (PC), while eigenvectors are the directions of those PCs. Essentially, eigenvectors define the new axes that represent the data with the highest variance, and eigenvalues quantify how much variance each eigenvector captures.


15.How do you decide the number of components to keep in PCA.

ans.You can choose the number of principal components that maximizes or optimizes your desired metric. In PCA, choose the smallest number of components that still capture most of the information in your data. A common approach is to pick enough components to cover about 95% of the total data variance

16. Can PCA be used for classification.

yes

17. What are the limitations of PCA.

ans. loss of interpretability, sensitivity to outliers and scale, and inability to capture non-linear relationships

18. How do KNN and PCA complement each other?

ans.KNN and PCA complement each other by combining dimensionality reduction with a distance-based classification technique. PCA reduces the complexity of the data by projecting it into a lower-dimensional space while preserving its variance, and KNN then uses this reduced data to classify new instances based on the similarity of the data points in the new space.

19. How does KNN handle missing values in a dataset.

ans.In KNN imputation: Each row is treated as a coordinate in a multi-dimensional space (each column represents a dimension). The algorithm calculates the distance between rows to identify the ones that are most similar. The missing value is then estimated based on the values of the closest rows.

20.What are the key differences between PCA and Linear Discriminant Analysis (LDA)?

ans.PCA and LDA are both dimensionality reduction techniques, but they differ significantly in their goals and approaches. PCA is an unsupervised technique that aims to find the principal components that capture the most variance in the data. LDA, on the other hand, is a supervised technique that focuses on maximizing the separation between classes or groups in the data.
