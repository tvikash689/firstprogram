3.Explain the concept of feature randomness in Random Forest.

ans.In Random Forest, feature randomness, also called feature bagging or the random subspace method, means that at each split in a decision tree within the forest, a random subset of the features is considered, not all features. This introduces more randomness into the model, leading to more diverse and less correlated trees, which generally improves the overall model performance.

20.What is the difference between Bagging and Boosting.


ans.Bagging and boosting are both ensemble methods that combine multiple models to improve machine learning performance, but they differ in how they train and combine these models. Bagging, or bootstrap aggregating, trains models independently in parallel, focusing on reducing variance and overfitting, while boosting trains models sequentially, each correcting the errors of the previous model and focusing on reducing bias.
