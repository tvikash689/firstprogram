7.What is Pre-Pruning in Decision Trees	


Pre-pruning in decision trees, also known as early stopping, is a technique that limits the growth of a decision tree before it fully develops. It prevents overfitting by halting the tree-building process based on predefined criteria, such as maximum tree depth or minimum impurity decrease.

8. What is Post-Pruning in Decision Trees	

ans.Post-pruning in decision trees is a technique where the decision tree is allowed to grow fully, potentially overfitting the training data, and then branches are removed to improve generalization performance on unseen data

9 What is the difference between Pre-Pruning and Post-Pruning	

ans.Pre-pruning and post-pruning are two methods for preventing overfitting in decision trees. Pre-pruning involves stopping the tree's growth before it reaches its maximum depth or size, while post-pruning allows the tree to grow fully and then removes branches that are not providing significant value

10.* What is a Decision Tree Regressor	

ans.A decision tree regressor is a supervised machine learning algorithm that uses a tree-like model to predict continuous numerical values by recursively splitting data based on features to minimize prediction errors, ultimately leading to leaf nodes that represent the predicted values. 

11.'* What are the advantages and disadvantages of Decision Trees	

ans.Decision trees offer advantages like interpretability and the ability to handle various data types, while also having disadvantages such as overfitting and instability. They are relatively easy to understand and visualize

12.* How does a Decision Tree handle missing values	

ans.The algorithm calculates impurity measures (like Gini impurity or entropy) while considering the weights of individual instances. This means instances with missing values are not excluded but are given less weight in the impurity calculation, effectively reducing their influence on the split decision

13.how does a Decision Tree handle categorical features	

Decision trees naturally handle categorical features by creating separate branches or splits in the tree for each distinct value within a categorical feature. This allows the tree to effectively partition the data based on the different categories, leading to more accurate predictions

14.What are some real-world applications of Decision Trees?

ans.Decision trees find applications in various fields like finance, healthcare, and marketing, aiding in tasks such as predicting customer behavior, risk assessment, and decision-making. They are also used in machine learning for classification and regression

26.Train a Gradient Boosting Classifier and visualize the ROC curve.

```python
#gradient classfier

from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split


from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix


import warnings
warnings.filterwarnings('ignore')
```

```python
X, y = make_classification(n_samples=1000, n_features=20, n_classes=2, random_state=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
clf = GradientBoostingClassifier()
clf
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
y_pred
```

```python
# Evaluate the model
print("Current model performance:")
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

27.Train an XGBoost Regressor and tune the learning rate using GridSearchCV.

```python
#hyperpraremeter tuning
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.05, 0.2]}
```

```python
param_grid
```

```python
gbc = GradientBoostingClassifier()
grid_search = GridSearchCV(estimator = gbc, param_grid=param_grid, cv = 5, verbose=3)
grid_search
```

```python
grid_search.fit(X_train, y_train)
```

```python
best_model = grid_search.best_estimator_
y_pred_tuned = best_model.predict(X_test)

# Evaluate the tuned model
print("Tuned model performance:")
print(f"Accuracy: {accuracy_score(y_test, y_pred_tuned)}")
print("Classification Report:")
print(classification_report(y_test, y_pred_tuned))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_tuned))
```
